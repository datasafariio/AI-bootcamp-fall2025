## 2. ğŸ” Loops (`for` and `while`)

```python
scores = [45, 67, 89, 23, 50, 76, 38]

pass_count = 0
for score in scores:
    if score >= 50:
        pass_count += 1

print(f"Number of students who passed: {pass_count}")
```

### âœ… Activity 2 Solution

```python
temperatures = [29, 32, 35, 28, 31, 33, 27]

# Using for loop
hot_days = 0
for t in temperatures:
    if t > 30:
        hot_days += 1
print(f"Hot days (for loop): {hot_days}")

# Using while loop
hot_days = 0
i = 0
while i < len(temperatures):
    if temperatures[i] > 30:
        hot_days += 1
    i += 1
print(f"Hot days (while loop): {hot_days}")
```

## 4. âš™ï¸ Functions

```python
def filter_passing(scores):
    passing = []
    for s in scores:
        if s >= 50:
            passing.append(s)
    return passing

scores = [45, 67, 89, 23, 50, 76, 38]
print("Passing scores:", filter_passing(scores))
```

## 5. ğŸš¨ Exception Handling

```python
data = ["12", "15", "not_available", "20"]

heights = []
for d in data:
    try:
        h = int(d)
        heights.append(h)
    except:
        print(f"Invalid entry skipped: {d}")

print("Valid tree heights:", heights)
```

# ğŸ† Capstone Mini Project: Agriculture & Climate Data

```python
yields = [1.2, 2.5, "error", 4.1, 5.0, 3.8, "missing", 1.7, 2.9]
rainfall = [400, 850, "NA", 1200, 950, 600, 300, "error", 700]

def clean_data(data_list):
    cleaned = []
    for d in data_list:
        try:
            cleaned.append(float(d))
        except:
            pass
    return cleaned

def classify_yields(yield_list):
    results = []
    for y in yield_list:
        if y < 2:
            results.append("Low")
        elif y <= 4:
            results.append("Medium")
        else:
            results.append("High")
    return results

def classify_rainfall(rainfall_list):
    results = []
    for r in rainfall_list:
        if r < 500:
            results.append("Low")
        elif r <= 1000:
            results.append("Normal")
        else:
            results.append("High")
    return results

def analyze_farm_data(yields, rainfall):
    clean_yields = clean_data(yields)
    clean_rainfall = clean_data(rainfall)
    
    y_classes = classify_yields(clean_yields)
    r_classes = classify_rainfall(clean_rainfall)
    
    print("Yield Classification:", y_classes.count("Low"), "Low,", y_classes.count("Medium"), "Medium,", y_classes.count("High"), "High")
    print("Rainfall Classification:", r_classes.count("Low"), "Low,", r_classes.count("Normal"), "Normal,", r_classes.count("High"), "High")

analyze_farm_data(yields, rainfall)
```

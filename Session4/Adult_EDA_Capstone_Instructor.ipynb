## Load and Inspect Data

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("../Data/Adult_cleaned.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

## Summary Statistics

```python

numeric_cols = ['age', 'education_num', 'hours_per_week']
df[numeric_cols].describe()
```

```python

print('Mean:\n',df[numeric_cols].mean())
print('Median:\n',df[numeric_cols].median())
print('Mode:\n',df[numeric_cols].mode())
print('Standard Deviation:\n',df[numeric_cols].std())
```

## Distribution Analysis

```python
# Visualizations for Age and Hours per Week

# Age Distribution
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['age'], kde=True)
plt.title("Distribution of Age")

# Hours per Week Distribution
plt.subplot(1,2,2)
sns.histplot(df['hours_per_week'], kde=True)
plt.title("Distribution of Hours per Week")
plt.tight_layout()
plt.show()
```

```python
# Boxplots for Age and Hours per Week

# Age Boxplot
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(x=df['age'])
plt.title("Boxplot of Age")

# Hours per Week Boxplot
plt.subplot(1,2,2)
sns.boxplot(x=df['hours_per_week'])
plt.title("Boxplot of Hours per Week")
plt.tight_layout()
plt.show()
```

## Categorical Analysis

```python
# Categorical Variable Visualizations

plt.figure(figsize=(14,4)) # Adjusted figure size for better visibility
plt.subplot(1,3,1) # First subplot
sns.countplot(x='sex', data=df) # Countplot
plt.title('Sex Distribution')
plt.subplot(1,3,2) # Second subplot
sns.countplot(y='education', data=df)
plt.title('Education Distribution')
plt.subplot(1,3,3) # Third subplot
sns.countplot(y='marital_status', data=df)
plt.title('Marital Status Distribution')
plt.tight_layout() # Adjust layout
plt.show()
```

```python

sns.countplot(x='sex', hue='income', data=df)
plt.title("Income by Gender")
plt.show()
```

## Scatterplots & Correlation

```python
plt.figure(figsize=(12,9))
# set the color of the regression line into yellow
sns.regplot(x='age', y='hours_per_week', data=df, scatter_kws={'alpha':0.3}, line_kws={'color': 'orange'})
plt.title("Age vs Hours per Week")
plt.show()
```

```python

# Correlation Matrix
df[['age','education_num','hours_per_week']].corr()
```

## Correlation Heatmap

```python

plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

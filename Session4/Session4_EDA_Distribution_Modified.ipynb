**Scenario:** You are a data analyst at a Tanzanian microfinance NGO managing loans for small farmers and entrepreneurs. For example, what's the average income? Is it skewed by a few high earners? This is where univariate statistics come in: they summarize one variable at a time using measures like mean, median, mode, variance, and standard deviation.



- The **univariate statistical measures** (mean, median, mode, variance, standard deviation) describe the **center** and **spread** of a single variable‚Äôs values. 
- The **mean** (average) sums all values and divides by count. 
- The **median** is the middle value that splits the sorted data in half (it is not influenced by extreme values). 
- The **mode** is the most frequent value. 
- **Variance** and **standard deviation** quantify spread: standard deviation measures the typical distance of data points from the mean. Together, these summarize a numeric field.

*2. Identify skewness and variance*
  - Scipy is a useful library for statistical computations
  - Install it via %pip install scipy if not already installed

```python
# import scipy.stats for skewness and variance calculations

from scipy import stats 
```

```python
# Select numeric columns
numeric_cols = df_soil.select_dtypes(include=[np.number]).columns
```

```python
# Initialize a dictionary to store variance values
variance_dict = {}
```

```python
# Use for loop to calculate skewness and variance for each numeric column

for col in numeric_cols:
    # Calculate skewness 
    skewness = stats.skew(df_soil[col].dropna())
    # Calculate variance
    variance = df_soil[col].var()
    # Store variance in dictionary
    variance_dict[col] = variance
    
    # Print results with interpretation
    print(f"\n{col}:")
    print(f"  Skewness: {skewness:.3f}", end="")
    
    if abs(skewness) < 0.5:
        print(" (Fairly symmetric)")
    elif skewness > 0:
        print(" (Right-skewed - tail on right)")
    else:
        print(" (Left-skewed - tail on left)")
    
    print(f"  Variance: {variance:.2f}")
```

<details>
<summary>Why Skewness Matters in Model Building</summary>

### Why Skewness Matters in Model Building
It is very important to investigate if the data is normally distributed or not before using the data to train your model. The following are the reasons why *Skewness* matters in model building.
1. **Model Performance & Accuracy**: Skewed data can mislead algorithms. Many machine learning models (linear regression, logistic regression, neural networks) assume data is approximately normally distributed.
If data is heavily skewed, the model may perform poorly on the majority class or underrepresent the minority class, leading to biased predictions.
2. **Outlier Sensitivity**: Right-skewed data has extreme high values. These outliers can disproportionately influence model coefficients and predictions.
Example: Income data with a few billionaires will skew the mean upward, misleading the model about typical income levels.
3. **Statistical Validity**: Many statistical tests assume normality. If your data is skewed, p-values and confidence intervals may be unreliable.
Example: If soil pH is right-skewed, a t-test comparing two soil samples may give incorrect results.
4. **Feature Scaling Issues**: Skewed features can cause problems during normalization/standardization. The mean and standard deviation become less representative.
This affects gradient descent in neural networks and distance-based algorithms (KNN, K-means).
5. **Class Imbalance (Classification)**: If your target variable is skewed (e.g., 95% positive, 5% negative), the model may simply predict the majority class and still appear "accurate."
Example: A fraud detection model that predicts "no fraud" for all transactions might achieve 95% accuracy but catch zero fraud cases!
</details>

<details>
<summary>What to do with Skewed Data</summary>

### What to do with skewed data
The following are common solutions for skewed data, they should be applied based on skewed feature dtype and skewness in the data before model training.

1. Log Transformation (for right-skewed data)
2. Square Root Transformation
3. Box-Cox Transformation (optimal)
4. Binning/Stratification (for classification with skewed target) by using `sklearn.utils class_weight`
5. Resampling (oversampling minority, undersampling majority)

***Bottom line***: Always inspect skewness during EDA. If detected, apply appropriate transformations before building your model to ensure better *accuracy* and *generalization*.

- You can learn more about *Skewness* here [Skewness](https://www.spss-tutorials.com/skewness/)

</details>

Key questions that can be asked and answered using a histogram of age distribution from the FinScope Tanzania dataset include:

- **What is the most common age group?** (Identify the mode or peak in the histogram.)
- **Is the age distribution symmetric, skewed left, or skewed right?** (Check the shape; e.g., a long tail on the right indicates older respondents are less common.)

2. üåä **KDE Plot**: Smoothed Distribution for continuous numerical data types

<summary>What is the key difference bewteen Histogram and KDE</summary>

- KDE is a smoothed version of the histogram

```python
# Visualization procedure

plt.figure(figsize=(8,5)) # Set figure size
sns.kdeplot(df_finscope['Age'], fill=True) # Create KDE plot for 'Age' column
plt.title("Age Density Curve ‚Äì FinScope Tanzania") # Set title
plt.xlabel("Age") # Set x-axis label
plt.ylabel("Density") # Set y-axis label
plt.show() # Display the KDE plot
```

<details>
<summary>How to interpret the KDE plot?</summary>
Key insights from the KDE plot for example age in the FinScope Tanzania dataset include:

1. **Shape and Symmetry**: The curve may appear roughly symmetric or slightly skewed, indicating if ages are evenly distributed around the mean or biased toward younger/older groups.
2. **Peaks (Modes)**: The highest point shows the most common age range (e.g., 30-40 years), highlighting dominant age groups in financial inclusion surveys.
3. **Spread and Density**: The width of the curve reveals the age range; narrower curves suggest concentrated ages, while wider ones indicate diversity.
4. **Tails**: Long tails on the right/left indicate skewness (e.g., fewer very young or very old respondents), potentially showing underrepresented groups.
5. **Comparison to Population**: If the curve peaks at working ages, it suggests the survey focuses on economically active individuals, useful for targeting financial services.
6. **Outliers**: Extreme ages (e.g., <18 or >80) appear as low-density tails, flagging potential data quality issues or rare cases.
These insights help assess demographic biases and inform policy on age-specific financial access.
</details>

<details>
<summary>What are the Key insights to consider when evaluating a boxplot?</summary>

Key insights to consider when evaluating a boxplot (e.g., for age in FinScope data):

1. **Median (Central Line)**: Indicates the middle value; compare to mean for skewness (e.g., if median < mean, right-skewed).
2. **Interquartile Range (IQR, Box)**: Shows 50% of data spread (Q3 - Q1); wider box means more variability.
3. **Whiskers**: Extend to min/max within 1.5√óIQR; length indicates data range excluding outliers.
4. **Outliers (Dots)**: Points beyond whiskers; flag potential errors or extreme values (e.g., very young/old ages).
5. **Symmetry**: If box is centered, data is symmetric; offset box suggests skew.
6. **Overall Spread**: Distance from whiskers to box shows tails; long whiskers indicate wide distribution.
7. **Comparisons**: For multiple boxplots (e.g., by gender), look for differences in medians or outliers to spot group variations.

These help detect skewness, outliers, and variability, guiding data cleaning or transformations. For example, many outliers in age might require capping or removal.

## üîç Bivariate Analysis (Scatterplots and Correlation) (25 minutes)

**Scenario:** To explore relationships, you examine pairs of numeric features. For example, is age related to loan amount? Does higher income correspond to larger loans?

**Technical:** A **scatter plot** shows one variable on the x-axis and another on the y-axis, with each point representing a record. Patterns (upward trend, cluster, none) may indicate **correlation**. We quantify this with the **Pearson correlation coefficient (r)**, which ranges from ‚Äì1 to +1. A value near +1 means a strong positive linear relationship; near 0 means no linear correlation. In Python, `df.corr()` computes pairwise correlations for numeric columns.

```python
# Scatter plot: Income vs. LoanAmount
plt.figure(figsize=(6,4))
plt.scatter(df['Income'], df['LoanAmount'], color='purple', alpha=0.7)
plt.title('Income vs. Loan Amount')
plt.xlabel('Income'); plt.ylabel('Loan Amount')
plt.show()

# Compute correlation
corr_matrix = df[['Age','Income','LoanAmount']].corr()
print(corr_matrix)
```

The scatter plot (`plt.scatter`) lets you visually assess if points cluster along a line. The printed correlation matrix shows numbers like: r(Income, LoanAmount) ‚âà 0.31, indicating a moderate positive correlation; r(Age, LoanAmount) might be negative. Remember: r is between ‚Äì1 and 1, measuring strength and direction of linear relation. A correlation of 0.7 or above is often considered strong; below 0.3 is weak.

* If points lie close to an upward-sloping line, r is positive (strong if near 1).
* If they form no clear pattern, r is near 0.
* A downward trend yields a negative r.

## üå°Ô∏è Correlation Heatmap (15 minutes)

**Scenario:** You have many numerical features and want an overview of all pairwise correlations to spot strong relationships.

**Technical:** A **heatmap** is a grid where cell colors represent values, useful for visualizing a correlation matrix. We will plot the correlation matrix from `df.corr()` using Seaborn.

```python
# Correlation heatmap
corr = df.corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

The heatmap above (with values annotated) quickly highlights high correlations: dark red (near +1) or dark blue (near ‚Äì1). For example, if Income and LoanAmount have a moderate positive correlation (~0.31), their cell will show ~0.31. Blocks of color show groups of features that move together. This multivariate view helps identify unexpected relationships or redundancy among features.
